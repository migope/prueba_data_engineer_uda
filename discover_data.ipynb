```python
with open('data/yelp_academic_dataset_review.json') as data_file:
    review = (json.loads(row) for row in data_file ) #the file is to big so we use a iterators
    review_cool = (row for row in review if row['cool'] != 0)
    display(review_cool.__next__())
```

```python
cool_values = dict()
with open('data/yelp_academic_dataset_review.json') as data_file:
    review = (json.loads(row) for row in data_file ) #the file is to big so we use a iterators
    for row in review:
        cool_score = row['cool']
        if cool_score in cool_values:
            cool_values[cool_score] = 1 + cool_values[cool_score]
        else:
            cool_values[cool_score] = 1
for i in range(-1, 503):
    print(f'score: {i}, count {cool_values[i]}')
```
